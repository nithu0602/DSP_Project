```python
from scipy.io import wavfile
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs,nithu = wavfile.read('/content/nithu.wav')
nithu = np.float32(nithu)
nithu=nithu.reshape(-1)
```

```python
n_acf=np.correlate(nithu,nithu,'full')
```

```python
plt.acorr(nithu, maxlags = 10)
```

```python
t=np.linspace(-10,10,529199)
plt.plot(t,n_acf)
plt.title("ACF nithya voice")
plt.grid()
```

```python
fs1,khrithi = wavfile.read('/content/khrithi.wav')
khrithi= np.float32(khrithi)
khrithi=khrithi.reshape(-1)
```

```python
k_acf=np.correlate(khrithi,khrithi,'full')
```

```python
plt.acorr(khrithi, maxlags = 10)
```

```python
t=np.linspace(-10,10,529199)
plt.plot(t,k_acf)
plt.title("ACF of khrithi voice")
plt.grid()
```
